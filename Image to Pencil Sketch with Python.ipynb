# Task - 4 Image to Pencil Sketch with Python

In this task, we need to read the image in RBG format and then convert it to a grayscale image. This will turn an image into a classic black and white photo. Then the next thing to do is invert the grayscale image also called negative image, this will be our inverted grayscale image. Inversion can be used to enhance details. Then we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image. This can be done by dividing the grayscale image by the inverted blurry image. Since images are just arrays, we can easily do this programmatically using the divide function from the cv2 library in Python.

**References:** https://thecleverprogrammer.com/2020/09/30/pencil-sketch-with-python/

# Importing the necessary libraries

```python
import cv2
import matplotlib.pyplot as plt
```

### Loading the image

```python
image = cv2.imread("/Users/saumyagupta/Desktop/LGM/istockphoto-810147810-612x612.jpeg")
```

### Converting the image to RGB format

* Matplotlib interprets images in RGB format, but OpenCV uses BGR format so to convert the image so that it's properly loaded, convert it before loading.

```python
image2 = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image2)
plt.axis(False)
plt.show()
```

### Converting the image to greyscale image

```python
grey_scale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(grey_scale_image)
```

### Inverting the image

```python
inverted_image = cv2.bitwise_not(grey_scale_image)
plt.imshow(inverted_image)
```

### Blurring the image

```python
blurred_image = cv2.GaussianBlur(inverted_image, (111,111), 0)
plt.imshow(blurred_image)
```

### Inverting the blurred image

```python
inverted_blur_image = cv2.bitwise_not(blurred_image)
plt.imshow(inverted_blur_image)
```

### Creating the pencil sketch using the grey scale and inverted image

```python
sketch_image = cv2.divide(grey_scale_image, inverted_blur_image, scale=256.0)
plt.imshow(sketch_image)
```

### Saving the image

```python
cv2.imwrite("sketch.png", sketch_image)
```

### Converting the image back to BRG format

```python
plt.figure(figsize = (10,10))
image3 = cv2.cvtColor(sketch_image, cv2.COLOR_RGB2BGR)
plt.imshow(image3)
plt.axis(False)
plt.show()
```
